## The aim of the manual and the scientific background

Semiconductor nanowires (NW) are highly promising candidates for the next generation of opto-electronic devices <a href="#footnote-1">[1]</a> mainly due to its ability to accommodate the strain developed during the heteroepitaxial growth. In order to understand and improve the entire device performance it is crucial to study the individual nanowire. Being that device operation might influence the structural properties(e.g.,heating, piezotronic/piezophototronic effect), both <i>ex situ</i> and <i>in situ</i> experiments are needed to further explore the influence of the strain and defects distribution on the (opto)electrical properties of the nanowires. 

X-ray diffraction is suitable for experiments in operando. With recent advances, X-ray diffraction techniques, such as coherent Bragg diffraction imaging (CBDI), now provide information about the strain and defects distribution within individual nanostructures  <a href="#footnote-2">[2]</a>. At the ESRF (beamline ID13) with the beam energy of 14.8 keV and a focused nano-beam of 200X200nm it has been shown that it is possible to map phases in a single nanowire. The NW of interest is identified and by fulfilling the Bragg condition for reflection, different structural phases a single NW were found. Setting the incoming X-ray beam angle to one of phases and scanning along the NW growth axis, it was possible to encode the axial phase structure of the single NW. 

On the other hand, scanning tunneling microscopy (STM) can be used to measure the electrical properties of individual nanowires on their own growth substrate <a href="#footnote-3">[3]</a>. Bellow are the steps for performing the measurements combings the two techniques both ex situ and in situ. The STM part of the experiment provided the insight into I-V characteristics of the NW and was performed before and after the X-ray experiment, providing the possibility to study the impact of x-ray radiation on the NW properties
 
The aim of this manual to help users how are trying to design the combined x-ray/STM experiments. In situ electrical measurements are performed in combination with X-ray diffraction by mounting a specifically designed STM onto the goniometer of a nanofocus beamline (BL). Here, the concept of the experiment is presented, aiming at the in situ characterization of Joule heating, piezoelectric response, or local strain induced by the nanoprobe, together with some initial results obtained on individual GaN nanowires.
The manual assumes user's familiarity with both, STM and x-ray imaging techniques. It is to be used as a guidance tools with the aim of making the BL time more efficient as the problems encountered are identified and possible ways around provided.

### Step 1. Aligning the STM proble to the NW of intrest

<strong>STM Imaging</strong>

For sample imaging the STM is operated in constant current mode. The tip is approached to the sample until the tunneling current is measured. Once the condition for imaging is satisfied (tunneling current) the tip is first retracted by 2um (total nw length) and then approached in steps of 20 nm until the tip and the top part of the NW are close enough to trigger the interaction for the feedback resulting in the Fig.2 a). 

<strong>Contacting a single NW with the STM tip</strong>

Once the NW of interest is identified, the tip is moved over the NW center (or until the largest z-piezo voltage is measured). Afterwards, the feedback is interrupted (see b)) and the current signal is provided with an external low noise current amplifier. External bias source is used to ramp the voltage (V) while measuring the current (I) such that I-V can be obtained over a specific NW. The contact is controlled with the tip z movement.

<strong>Note 2:</strong>
Unfortunately the W tip cannot be used during in-situ without the additional source of nitrogen gas where one would enclose the setup in nitrogen environment. This is because the W tip oxidizes heavily during the exposure to the x-rays rendering imaging/contacting impossible. Hence, expensive and not as sharp PtIr probes are required. In addition, the best practice is to carry a standard sample for the tip characterization and mount it close to the sample of interest, making the time more efficient.

## Solution to the problems accountered

1. The photoejected electrons from the sample are interfering with the measurements of tunneling current, thus it is necessary to reduce the intensity of the beam by changing the attenuation. It is highly advised to be able to change the attenuation remotely in order to prevent the entrance the the hutch and thus switching the beam on and off contantly.


2. STM instabilities are best explained by the stage tilt drift due to the large angle (23ged) required for Bragg reflection. If it is not possible to used a higher energy beam where the Bragg reflection is smaller perhaps moving to GaN [0001] Bragg angle is sufficient.

3. The tip must be stable, firmly glued to the tip stage! The best practice is to mount several probes in advance allowing the conductive glue to harden.
